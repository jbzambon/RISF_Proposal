```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset

import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
```

```python
file = '/mnt/home/jbzambon/zeljko/z_0.nc'
# file = '/mnt/home/jbzambon/zeljko/z_50.nc'
# file = '/mnt/home/jbzambon/zeljko/z_250.nc'
# file = '/mnt/home/jbzambon/zeljko/z_500.nc'
nc = Dataset(file)
```

```python
file = '/mnt/home/jbzambon/zeljko/z_0.nc'
nc = Dataset(file)
lat = nc['lat_rho'][:,0]
lon = nc['lon_rho'][0,:]
u = nc['u'][:,:]
v = nc['v'][:,:]
mag = ( u**2 + v**2 ) **0.5

fig=plt.figure(frameon=True,figsize=(15,20))
[X,Y] = np.meshgrid(lon,lat)
parallels = np.arange(0.,90,10.)
meridians = np.arange(180.,360.,10.)

map = Basemap(projection='merc',
      resolution='i',lat_0=((np.max(lat)-np.min(lat))/2),
      lon_0=((np.max(lon)-np.min(lon))/2),
      llcrnrlon=np.min(lon),llcrnrlat=np.min(lat),
      urcrnrlon=np.max(lon),urcrnrlat=np.max(lat))
map.drawcoastlines()
map.drawcountries()
map.drawstates()
map.pcolormesh(X,Y,np.squeeze(mag)[:,:],cmap='jet',vmin=0,vmax=2.5,latlon='true')
map.drawparallels(parallels,labels=[1,0,0,0],fontsize=18)
map.drawmeridians(meridians,labels=[0,0,0,1],fontsize=18)

# cbar=map.colorbar(location='bottom',ticks=np.arange(0,2.51,0.5))

fig.savefig('zeljko_0.png',facecolor='w',format='png',transparent=False)
print(np.max(np.max(mag)))
```
